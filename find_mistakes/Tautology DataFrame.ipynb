## Преобразование данных КРУТа в удобный формат

```python
import pandas as pd
```

```python
data, tags = pd.read_csv('data2.csv'), pd.read_csv('tags.csv')
data.head()
```

```python
tags.head()
```

```python
tags = pd.DataFrame(tags.groupby('annotator_sentence.id')['tag'].apply(set))
tags.head()
```

```python

def read_data(data, tags):
    tautos_df = pd.merge(data,tags, how='outer',on='annotator_sentence.id')
    tautos_df = tautos_df.loc[tautos_df['annotated'] == 1]
    tautos_df.index=range(51365)
    return tautos_df
tautos_df_new=read_data(data, tags)
```

```python
tautos_df_new.head()
```

```python
tautos = []
for v in tautos_df_new['tag']:
    if isinstance(v, set) and ('tauto' in v):
        tautos.append(1.0)
    else:
        tautos.append(0.0)
tautos_new=pd.Series(tautos)
```

```python
tautos_df_new.loc[:,'tautos']=tautos_new
tautos_df_new.head()
```

```python
tautos_df_new.to_csv('tautos_df.csv')
```
